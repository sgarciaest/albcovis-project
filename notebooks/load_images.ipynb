# Loading images locally

```python
local_image = "../data/source/images/mbid-0d4aa90e-02d9-3003-a77f-a9706c1ac77e-40654268451_thumb500.jpg"
```

```python
from PIL import Image

# Load image from local file
image = Image.open(local_image)
image.show()  # Optional: displays the image
print(type(image))
```

```python
import cv2

# Load image in color
image = cv2.imread(local_image, cv2.IMREAD_COLOR)
cv2.imshow("Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

If using a notebook, better use `matplotlib` to show the images. Notebooks can't handle properly GUI windows.

```python
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open(local_image)

plt.imshow(img)
# plt.axis("off")
# plt.title("Album Cover")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load image
image = cv2.imread(local_image)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for proper color

# Display with matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Hide axes
plt.title("Image")
plt.show()
```

# Loading images from URL

```python
import requests
from PIL import Image
from io import BytesIO

url = "https://coverartarchive.org/release/0d4aa90e-02d9-3003-a77f-a9706c1ac77e/40654268451-500.jpg"
response = requests.get(url)
image = Image.open(BytesIO(response.content))
plt.imshow(image)
# plt.axis("off")
# plt.title("Album Cover")
plt.show()
```

`io` is a built-in Python module (no installation needed) that provides tools for handling in-memory file-like objects.

`BytesIO` specifically:
- Think of it as a virtual file stored in RAM, not on disk.
- It allows you to treat raw bytes (like image data from a web request) as if they were a real file.

When you fetch an image from a URL using `requests.get()`, the image data you get is binary data  just a long stream of bytes.

But `PIL.Image.open()` expects a file or file-like object  not just bytes.

```python
from urllib.parse import urlparse
from pathlib import Path


url = "https://coverartarchive.org/release/0d4aa90e-02d9-3003-a77f-a9706c1ac77e/40654268451-500.jpg"

parsed_url = urlparse(url)
filename = Path(parsed_url.path).name or "downloaded_image"
print(filename)
print(type(filename))
```
